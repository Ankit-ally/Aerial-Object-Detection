```python
#Import Libraries
import numpy as np
import pandas as pd
```

```python
# Load Dataset
df = pd.read_csv("/content/ZOMATO.csv")

df.head()
```

```python
# DataSet Shape
df.shape
```

```python
# Column Overview
df.columns
```

```python
# =========================================================
# 3. DATA OVERVIEW
# =========================================================

print("Shape of dataset:", df.shape)
print("\nColumns:")
print(df.columns)

print("\nInfo:")
df.info()
```

```python
# =========================================================
# 4. DESCRIPTIVE STATISTICS — OVERALL SUMMARY
# =========================================================
df.describe()
```

```python

# =========================================================
# 5. MEASURES OF CENTRAL TENDENCY
# =========================================================

print("Mean Price:", df["Price"].mean())
print("Median Price:", df["Price"].median())
print("Mode Price:", df["Price"].mode()[0])

print("\nMean Rating:", df["Avg ratings"].mean())
print("Median Rating:", df["Avg ratings"].median())
print("Mode Rating:", df["Avg ratings"].mode()[0])

```

```python

# =========================================================
# 6. MEASURES OF DISPERSION
# =========================================================

print("Price Variance:", df["Price"].var())
print("Price Standard Deviation:", df["Price"].std())
print("Price Range:", df["Price"].max() - df["Price"].min())

print("\nDelivery Time Variance:", df["Delivery time"].var())
print("Delivery Time Std Dev:", df["Delivery time"].std())
print("Delivery Time Range:", df["Delivery time"].max() - df["Delivery time"].min())

```

```python

# =========================================================
# 7. QUARTILES AND PERCENTILES
# =========================================================

print("Quartiles for Price:")
print(df["Price"].quantile([0.25,0.5,0.75]))

print("\nPercentiles for Price:")
print(np.percentile(df["Price"], [10,25,50,75,90]))
```

```python

# =========================================================
# 8. MINIMUM AND MAXIMUM VALUES
# =========================================================

print("Minimum Price:", df["Price"].min())
print("Maximum Price:", df["Price"].max())

print("Minimum Rating:", df["Avg ratings"].min())
print("Maximum Rating:", df["Avg ratings"].max())


```

```python

# =========================================================
# 9. COUNT STATISTICS
# =========================================================

print("Total Restaurants:", len(df))
print("Unique Cities:", df["City"].nunique())
print("Most Common City:")
print(df["City"].value_counts().head())
```

```python


# =========================================================
# 10. GROUPED DESCRIPTIVE STATISTICS
# =========================================================

print("Average Price per City:")
print(df.groupby("City")["Price"].mean().sort_values(ascending=False))

print("\nAverage Rating per City:")
print(df.groupby("City")["Avg ratings"].mean().sort_values(ascending=False))
```

```python


# =========================================================
# 11. MULTIPLE AGGREGATIONS
# =========================================================

df.groupby("City").agg({
    "Price":["mean","min","max"],
    "Avg ratings":["mean","count"],
    "Delivery time":["mean"]
})
```

```python


# =========================================================
# 12. CORRELATION BETWEEN NUMERICAL FEATURES
# =========================================================

df[["Price","Avg ratings","Delivery time","Total ratings"]].corr()


```

```python
# =========================================================
# 13. DISTRIBUTION INSIGHT QUESTIONS (FOR STUDENTS)
# =========================================================

print("Average Price:", df["Price"].mean())
print("Median Price:", df["Price"].median())

if df["Price"].mean() > df["Price"].median():
    print("Distribution is Right Skewed")
else:
    print("Distribution is Left Skewed")

```

```python

# =========================================================
# 14. QUICK INSIGHT SUMMARY
# =========================================================

print("Top 5 Highest Rated Restaurants")
df.sort_values("Avg ratings", ascending=False).head()

```

```python
# 📊 DISTRIBUTION ANALYSIS + VISUALIZATION MODULE
```

We already summarized data using numbers.
But numbers don’t always reveal patterns.
To truly understand data, we must visualize its distribution.

** What is Distribution? **
Distribution shows how values are spread across a dataset.

```python
#If marks of students are plotted, distribution shows whether most students scored high or low.
```

```python
# Step 1 — Import Visualization Libraries
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

Step 2 — Histogram (Most Important Distribution Plot)
Concept Explanation

Histogram shows:

frequency

spread

shape

```python
plt.figure(figsize=(8,5))
plt.hist(df["Price"], bins=20)
plt.title("Price Distribution")
plt.xlabel("Price")
plt.ylabel("Number of Restaurants")
plt.show()
```

Step 3 — Distribution Plot (Smooth Curve)

Shows probability density.

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Price"], kde=True)
plt.title("Price Distribution with Density Curve")
plt.show()
```

📊 Step 4 — Distribution of Ratings

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Avg ratings"], bins=15, kde=True)
plt.title("Ratings Distribution")
plt.show()
```

📦 Step 5 — Boxplot (Detect Outliers)
Explain Concept First

Boxplot summarizes distribution using 5 values:
min, Q1, median, Q3, max.

```python
plt.figure(figsize=(7,4))
sns.boxplot(x=df["Price"])
plt.title("Boxplot of Price")
plt.show()
```

Step 6 — Compare Distributions by Category

Example → Price distribution per city.

```python
plt.figure(figsize=(12,6))
sns.boxplot(x="City", y="Price", data=df)
plt.xticks(rotation=90)
plt.title("Price Distribution Across Cities")
plt.show()
```

Step 7 — Count Plot (Categorical Distribution)

Shows frequency of categories.

```python
plt.figure(figsize=(10,5))
sns.countplot(x="City", data=df, order=df["City"].value_counts().index)
plt.xticks(rotation=90)
plt.title("Number of Restaurants per City")
plt.show()
```

📈 Step 8 — Distribution of Delivery Time

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Delivery time"], bins=20, kde=True)
plt.title("Delivery Time Distribution")
plt.show()
```

📉 Step 9 — Skewness Concept
Definition

Skewness tells whether distribution is symmetric or tilted.

```python
df["Price"].skew()
```

📊 Step 10 — Compare Mean vs Median Visually

```python
print("Mean:", df["Price"].mean())
print("Median:", df["Price"].median())
```

Explain:

If mean > median → right skew
If mean < median → left skew
